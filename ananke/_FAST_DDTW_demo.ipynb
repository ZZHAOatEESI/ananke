## test case 2: run time
* Complexity: $O(n)$

```python
from time import time
```

```python
signal_3 = np.repeat(signal_1, 20)
signal_4 = np.repeat(signal_2, 20)
```

```python
# plot raw signal
plot_raw_signals(signal_3, signal_4)
```

```python
# fast ddtw
start = time()
distance, path = fast_ddtw(signal_3, signal_4)
end = time()
runtime_fastddtw = end - start
print('runtime:',  runtime_fastddtw)
print('distance:', distance)
```

```python
# plot path
plot_alignment_path(path)
```

```python
# plot aligned signal
plot_aligned_signals(signal_3, signal_4, path)
```
